# Numerical Variables

```python
numerical_features=[feature for feature in dataset.columns if dataset[feature].dtypes!='O']
numerical_features
```

```python
dataset[numerical_features].head()
```

# Temporal Variables(Eg: Datetime Variables)

```python
year_feature = [feature for feature in numerical_features if 'auto_year' in feature]

year_feature
```

```python
# let's explore the content of these year variables
for feature in year_feature:
    print(feature, dataset[feature].unique())
```

```python
for feature in year_feature:
    data=dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

```python
## Lets analyze the Temporal Datetime Variables
## We will check whether there is a relation between year the house is sold and the sales price


dataset.groupby('auto_year')['Fraud_0_1'].median().plot()
plt.xlabel('auto_year')
plt.ylabel('Fraud_0_1')
plt.title("auto_year vs YearSold")
```

*********** WE COULD SEE I THE HISTOGRAM FOR THE VARIABLE "auto_year" THERE IS SKEWNESS PROBLEM WE COULD NOT DETERMINE IT
IT SHOULD GO THROUGH TRANSFORMATION ****************

```python
## Here we will compare the difference between All years feature with SalePrice

for feature in year_feature:
    
    
        plt.scatter(dataset['auto_year'],dataset['Fraud_0_1'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
discrete_feature=[feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_feature+['Id']]

print("Discrete Variables Count: {}".format(len(discrete_feature)))
```

```python
dataset[discrete_feature].head()
```

```python
## Lets Find the realtionship between them and Sale PRice

for feature in discrete_feature:
    data=dataset.copy()
    data.groupby(feature)['Fraud_0_1'].mean().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('Fraud_0_1')
    plt.title(feature)
    plt.show()
```

# Continuous Variable

```python
continuous_feature=[feature for feature in numerical_features if feature not in discrete_feature+year_feature]

print("Continuous feature Count {}".format(len(continuous_feature)))
```

```python
dataset[continuous_feature].head()
```

```python
dataset[continuous_feature].hist(figsize=(15,15), xrot=-45, bins=10) ## Display the labels rotated by 45 degress

# Clear the text "residue"
plt.show()
```

```python
for feature in continuous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data[feature].hist(figsize=(3,3), xrot=-45, bins=10)  ## Display the labels rotated by 45 degress
        # Clear the text "residue"
        plt.show()

```

# Handling Outliers

```python
for feature in continuous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()          
    
```
